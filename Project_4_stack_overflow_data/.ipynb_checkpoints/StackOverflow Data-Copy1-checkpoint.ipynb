<b>Question 2) How has Gender changed over time?</b>

```python
#Count number of gender and store in a dict
df_ext_2018_gender = df_ext_2018['Gender'].value_counts()
df_ext_2017_gender = df_ext_2017['Gender'].value_counts()
df_ext_2016_gender = df_ext_2016['Gender'].value_counts()
df_ext_2015_gender = df_ext_2015['Gender'].value_counts()
df_ext_2014_gender = df_ext_2014['Gender'].value_counts()
```

```python
#Calculate proportion of males and females in 2018
#num_males = df_ext_2018['Gender'] = 'Male'
num_males_2018 = df_ext_2018['Gender'].loc[df_ext_2018['Gender']=='Male'] #find number of males in Gender column
num_females_2018 = df_ext_2018['Gender'].loc[df_ext_2018['Gender']=='Female'] #find number of females in Gender column
male_female_tot_2018 = len(num_males_2018) + len(num_females_2018) #sum up total 

males_prop_2018 = len(num_males_2018)/male_female_tot_2018 #calcuate proportion of males
females_prop_2018 = len(num_females_2018)/male_female_tot_2018 #calculate proportion of females

#Create new dataframe of just males and females for plotting purposes
male_female_tot_list_2018 = [[males_prop_2018], [females_prop_2018]] #create list to store in dataframe
male_female_total_df_2018 = pd.DataFrame(male_female_tot_list_2018, columns=['Total'])
male_female_total_df_2018.index.names = ['Gender']

male_female_total_df_2018.rename(index={0:'Males', 1:'Females'},inplace=True) #replace index vales with Gender names for plotting
```

```python
print('The proportion of males is: ', males_prop_2018*100, '%', sep='')
print('The proportion of females is: ', females_prop_2018*100, '%', sep='')
```

```python
#Plot
male_female_total_df_2018.plot.bar(figsize=(20,10), title='Gender in 2018', rot=45);
```

```python
#Calculate proportion of males and females in 2017

num_males_2017 = df_ext_2017['Gender'].loc[df_ext_2017['Gender']=='Male'] #find number of males in Gender column
num_females_2017 = df_ext_2017['Gender'].loc[df_ext_2017['Gender']=='Female'] #find number of females in Gender column
male_female_tot_2017 = len(num_males_2017) + len(num_females_2017) #sum up total 

males_prop_2017 = len(num_males_2017)/male_female_tot_2017 #calcuate proportion of males
females_prop_2017 = len(num_females_2017)/male_female_tot_2017 #calculate proportion of females

#Create new dataframe of just males and females for plotting purposes
male_female_tot_list_2017 = [[males_prop_2017], [females_prop_2017]] #create list to store in dataframe
male_female_total_df_2017 = pd.DataFrame(male_female_tot_list_2017, columns=['Total'])
male_female_total_df_2017.index.names = ['Gender']

male_female_total_df_2017.rename(index={0:'Males', 1:'Females'},inplace=True) #replace index vales with Gender names for plotting
```
